```python
import sys, time
import sevseg

class Timer:
    def __init__(self, seconds_left):
        self.seconds_left = seconds_left
        
    def update_time(self):
        self.hours = str(self.seconds_left // 3600)
        self.minutes = str((self.seconds_left % 3600) // 60)
        self.seconds = str(self.seconds_left % 60)
        
    def display_time(self):
        # Generic display, subclasses will override this
        print(f"{self.hours}:{self.minutes}:{self.seconds}")
        
    def run(self):
        try:
            while self.seconds_left > 0:
                self.update_time()
                self.display_time()
                print("\nPress Ctrl+C to exit")
                time.sleep(1)
                self.seconds_left -= 1
            print("\nTime's up!")
        except KeyboardInterrupt:
            print("Timer stopped.")
            sys.exit()
            
class SevenSegmentDisplayTimer(Timer):
    def display_time(self):
        hDigits = sevseg.getSevSegStr(self.hours, 2)
        hTopRow, hMiddleRow, hBottomRow = hDigits.splitlines()

        mDigits = sevseg.getSevSegStr(self.minutes, 2)
        mTopRow, mMiddleRow, mBottomRow = mDigits.splitlines()

        sDigits = sevseg.getSevSegStr(self.seconds, 2)
        sTopRow, sMiddleRow, sBottomRow = sDigits.splitlines()

        # Display the digits
        print('\n' * 60)
        print(hTopRow    + '     ' + mTopRow    + '     ' + sTopRow)
        print(hMiddleRow + '  *  ' + mMiddleRow + '  *  ' + sMiddleRow)
        print(hBottomRow + '  *  ' + mBottomRow + '  *  ' + sBottomRow)
        
# Initialize the timer with 1 hour (3600 seconds)
seven_segment_timer = SevenSegmentDisplayTimer(30)
seven_segment_timer.run()

```
