```python
import random

class DataModel:
    def __init__(self):
        self.object_pronouns = ['Her', 'Him', 'Them']
        self.possessive_pronouns = ['Her', 'His', 'Their']
        self.personal_pronouns = ['She', 'He', 'They']
        self.states = ['California', 'Texas', 'Florida', 'New York', 'Pennsylvania',
                       'Illinois', 'Ohio', 'Georgia', 'North Carolina', 'Michigan']
        self.nouns = ['Athlete', 'Clown', 'Shovel', 'Paleo Diet', 'Doctor', 'Parent',
                      'Cat', 'Dog', 'Chicken', 'Robot', 'Video Game', 'Avocado',
                      'Plastic Straw', 'Serial Killer', 'Telephone Psychic']
        self.places = ['House', 'Attic', 'Bank Deposit Box', 'School', 'Basement',
                       'Workplace', 'Donut Shop', 'Apocalypse Bunker']
        self.when = ['Soon', 'This Year', 'Later Today', 'RIGHT NOW', 'Next Week']

class HeadlineGenerators:
    def __init__(self, data_model):
        self.data = data_model

    def generate_are_millennials_killing_headline(self):
        noun = random.choice(self.data.nouns)
        return f'Are Millennials Killing the {noun} Industry?'

    def generate_what_you_dont_know_headline(self):
        noun = random.choice(self.data.nouns)
        pluralNoun = f'{random.choice(self.data.nouns)}s'
        when = random.choice(self.data.when)
        return f'Without This {noun}, {pluralNoun} Could Kill You {when}'

    def generate_big_companies_hate_her_headline(self):
        pronoun = random.choice(self.data.object_pronouns)
        state = random.choice(self.data.states)
        noun1 = random.choice(self.data.nouns)
        noun2 = random.choice(self.data.nouns)
        return f'Big Companies Hate {pronoun}! See How This {state} {noun1} Invented a Cheaper {noun2}'

    def generate_you_wont_believe_headline(self):
        state = random.choice(self.data.states)
        noun = random.choice(self.data.nouns)
        pronoun = random.choice(self.data.possessive_pronouns)
        place = random.choice(self.data.places)
        return f'You Won\'t Believe What This {state} {noun} Found in {pronoun} {place}'

    def generate_dont_want_you_to_know_headline(self):
        pluralNoun1 = f'{random.choice(self.data.nouns)}s'
        pluralNoun2 = f'{random.choice(self.data.nouns)}s'
        return f'What {pluralNoun1} Don\'t Want You To Know About {pluralNoun2}'

    def generate_gift_idea_headline(self):
        number = random.randint(7, 15)
        noun = random.choice(self.data.nouns)
        state = random.choice(self.data.states)
        return f'{number} Gift Ideas to Give Your {noun} From {state}'

    def generate_reasons_why_headline(self):
        number1 = random.randint(3, 19)
        pluralNoun = f'{random.choice(self.data.nouns)}s'
        number2 = random.randint(1, number1)  # Ensure number2 is no larger than number1
        return f'{number1} Reasons Why {pluralNoun} Are More Interesting Than You Think (Number {number2} Will Surprise You!)'

    def generate_job_automated_headline(self):
        state = random.choice(self.data.states)
        noun = random.choice(self.data.nouns)
        i = random.randint(0, 2)
        pronoun1 = self.data.possessive_pronouns[i]
        pronoun2 = self.data.personal_pronouns[i]
        if pronoun1 == 'Their':
            return f'This {state} {noun} Didn\'t Think Robots Would Take {pronoun1} Job. {pronoun2} Were Wrong.'
        else:
            return f'This {state} {noun} Didn\'t Think Robots Would Take {pronoun1} Job. {pronoun2} Was Wrong.'


class ClickbaitGenerator:
    def __init__(self):
        self.data_model = DataModel()
        self.headline_generators = HeadlineGenerators(self.data_model)

    def run(self):
        print("Clickbait headlines")
        print('Our website needs to trick people into looking at ads!')

        while True:
            response = input("Enter the number of headlines to generate = ")
            if not response.isdecimal():
                print("Please enter a number")
                continue
            
            number_of_headlines = int(response)
            break

        for _ in range(number_of_headlines):
            # Randomly choose a headline generation method to call
            method = random.choice([
                self.headline_generators.generate_are_millennials_killing_headline,
                self.headline_generators.generate_what_you_dont_know_headline,
                self.headline_generators.generate_big_companies_hate_her_headline,
                self.headline_generators.generate_you_wont_believe_headline,
                self.headline_generators.generate_dont_want_you_to_know_headline,
                self.headline_generators.generate_gift_idea_headline,
                self.headline_generators.generate_reasons_why_headline,
                self.headline_generators.generate_job_automated_headline
            ])
            print(method())
        print()

        website = random.choice(['wobsite', 'blag', 'Facebuuk', 'Googles', 'Facesbook', 'Tweedie', 'Pastagram'])
        when = random.choice(self.data_model.when).lower()
        print(f'Post these to our {website} {when}, or you\'re fired!')
if __name__ == '__main__':
    clickbait_generator = ClickbaitGenerator()
    clickbait_generator.run()

```
